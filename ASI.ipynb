```python
ans = list() #Здесь будут ответы
```

## Принцип работы

Сперва переводим все фотографии в формат RGB для удобства работы. После происходит вызов функции get_person_name. Она возвращает id пользователя на leader-id. На фотографии находится лицо, вырезается, отправляется в нейронную сеть, которая возвращает 128 числовых признаков(дескрипторов), после чего ищется кратчайщее евклидово расстояние. 

## Что могло пойти не так?

Если на фотографии нет лица, то все будет ок. Если человека на фотографии нет в базе, всегда может найтись похожий человек, с которым программа может его спутать. Также при поиске евклидова расстояния разница между двумя людьми может быть одной миллионной(допустим), на тесте мы выводили одного человека, а так мы выводим нескольких наиболее похожих людей.

```python
for people in people_files:
    img = color.gray2rgb(io.imread(people))
    tens = str(people[58:]) + " " + str(get_person_name(img))
    ans.append(tens)
    print(tens)
```

```python
#Сохраняем в формате json
with open("asi_ans.json", 'w') as file:
    file.write(dumps(ans, indent=4))
    file.close()
```
